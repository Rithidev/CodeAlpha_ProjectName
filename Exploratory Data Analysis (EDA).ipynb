```python

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
salary_nums = df['Salary Estimate'].str.extract(r'\$(\d+)K-\$(\d+)K')
df['AvgSalary'] = (pd.to_numeric(salary_nums[0]) + pd.to_numeric(salary_nums[1])) / 2

plt.figure(figsize=(10,6))
sns.histplot(df['AvgSalary'], bins=20, kde=True)
plt.title("Average Salary Distribution")
plt.xlabel("Avg Salary (K USD)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))
df['Industry'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Industries Hiring Business Analysts")
plt.xlabel("Industry")
plt.ylabel("Job Count")
plt.show()
```

```python
plt.figure(figsize=(20,10))
df['Rating'].dropna().hist()
plt.title("Company Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['Location'].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Locations by Job Count")
plt.xlabel("Location")
plt.ylabel("Number of Jobs")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,8))
df.groupby("Size")["AvgSalary"].mean().dropna().sort_values(ascending=False).plot(kind="bar")
plt.title("Company Size vs Avg Salary")
plt.xlabel("Company Size")
plt.ylabel("Avg Salary (K USD)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(df['Rating'], df['AvgSalary'])
plt.title("Rating vs Average Salary")
plt.xlabel("Rating")
plt.ylabel("Avg Salary (K USD)")
plt.tight_layout()
plt.show()

```
